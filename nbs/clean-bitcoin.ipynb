```python
DOW_DATA = '../data/dow-data.csv'
EUR_DATA = '../data/eur-data.csv'
GOLD_DATA = '../data/gold-data.csv'
MIN_BIT = '../data/bitcoin-min.csv'
```

```python
bitcoin = pd.read_csv(MIN_BIT)
```

```python
bitcoin.head()
```

```python
import datetime
bitcoin['Date'] = bitcoin['Timestamp'].apply(lambda x: datetime.datetime.fromtimestamp(x).strftime('%Y-%m-%d %H:%M:%S'))
```

```python
bitcoin.drop('Timestamp',axis=1, inplace=True)
```

```python
bitcoin = bitcoin.set_index(pd.DatetimeIndex(bitcoin['Date']))
```

```python
bitcoin.head()
```

```python
open = bitcoin.Open.resample('D').first()
high = bitcoin.High.resample('D').max()
low = bitcoin.Low.resample('D').min()
close = bitcoin.Close.resample('D').last()
vol = bitcoin["Volume_(BTC)"].resample('D').sum()
vol_cur = bitcoin["Volume_(Currency)"].resample('D').sum()
weight = bitcoin.Weighted_Price.resample('D').last()
```

```python
bitcoin_day = pd.concat([open,high,low,close,vol,vol_cur,weight], axis=1)
```

```python
bitcoin_day
```

```python
bitcoin_day.to_csv('../data/bitcoin-day.csv')
```
