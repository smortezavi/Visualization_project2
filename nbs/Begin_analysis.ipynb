## Fixing commas and floats

```python
columns_with_commas = ['dow_price', 'dow_open', 
                       'dow_high', 'dow_low', 
                       'gold_price', 'gold_open', 
                       'gold_high', 'gold_low']
```

```python
for cols in columns_with_commas:
    final_data[cols] = final_data[cols].map(lambda x:float(x.split(',')[0]+x.split(',')[1]))
```

## dividing by max to get a relative feel 

```python
final_data['dow_price_soft_max'] = (final_data.dow_price / final_data.dow_price.max())
final_data['gold_price_soft_max'] = (final_data.gold_price / final_data.gold_price.max())
final_data['eur_price_soft_max'] = (final_data.eur_price / final_data.eur_price.max())
final_data['bit_price_soft_max'] = (final_data.bit_price / final_data.bit_price.max())
```

```python
final_data['bit_price_log'] = np.log(final_data.bit_price)
```

```python
final_data['bit_price_log_norm'] = (final_data.bit_price_log / final_data.bit_price_log.max())
```

```python
final_data.sort_values(by = ['Date']).transpose()
```

## Plots

** In this project we will investigate bitcoins role in the world market**

```python
df = final_data
data = [go.Scatter(x = df.Date, y = df.dow_price)]
py.iplot(data, filename = 'DOW')
```

## Non-normalized plots

```python
dow_ = go.Scatter(
            x = df.Date,
            y = df.dow_price,
            name = "DOW",
            line = dict(color = '#17BECF'),
            opacity = 0.8)

gold_ = go.Scatter(
                x = df.Date,
                y = df.gold_price,
                name = "GOLD",
                line = dict(color = '#17BECF'),
                opacity = 0.8)

euro_ = go.Scatter(
                x = df.Date,
                y = df.eur_price,
                name = "EURO",
                line = dict(color = '#7F7F7F'),
                opacity = 0.8)

bit_ = go.Scatter(
                x = df.Date,
                y = df.bit_price,
                name = "BIT-COIN",
                line = dict(color = 'green'),
                opacity = 0.8)

data = [dow_, gold_, euro_, bit_]

layout = dict(
    title = "Markets",
    xaxis = dict(
        range = ['2015-01-12','2018-03-26'])
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename = "Manually Set Range")
```

## Normalized Plots

```python
dow_norm = go.Scatter(
            x = df.Date,
            y = df.dow_price_soft_max,
            name = "DOW",
            line = dict(color = '#17BECF'),
            opacity = 0.8)

gold_norm = go.Scatter(
                x = df.Date,
                y = df.gold_price_soft_max,
                name = "GOLD",
                line = dict(color = '#17BECF'),
                opacity = 0.8)

euro_norm = go.Scatter(
                x = df.Date,
                y = df.eur_price_soft_max,
                name = "EURO",
                line = dict(color = '#7F7F7F'),
                opacity = 0.8)

bit_norm = go.Scatter(
                x = df.Date,
                y = df.bit_price_soft_max,
                name = "BIT-COIN",
                line = dict(color = 'green'),
                opacity = 0.8)

data = [dow_norm, gold_norm, euro_norm, bit_norm]

layout = dict(
    title = "Markets",
    xaxis = dict(
        range = ['2015-01-12','2018-03-26'])
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename = "Manually Set Range")
```

** We further log the bit prices and then divide by max to get a better picture of the different markets **

```python
dow_norm = go.Scatter(
            x = df.Date,
            y = df.dow_price_soft_max,
            name = "DOW",
            line = dict(color = '#17BECF'),
            opacity = 0.8)

gold_norm = go.Scatter(
                x = df.Date,
                y = df.gold_price_soft_max,
                name = "GOLD",
                line = dict(color = '#17BECF'),
                opacity = 0.8)

euro_norm = go.Scatter(
                x = df.Date,
                y = df.eur_price_soft_max,
                name = "EURO",
                line = dict(color = '#7F7F7F'),
                opacity = 0.8)

bit_norm_log = go.Scatter(
                x = df.Date,
                y = df.bit_price_log_norm,
                name = "BIT-COIN",
                line = dict(color = 'green'),
                opacity = 0.8)

data = [dow_norm, gold_norm, euro_norm, bit_norm_log]

layout = dict(
    title = "Markets",
    xaxis = dict(
        range = ['2015-01-12','2018-03-26'])
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename = "Manually Set Range")
```
