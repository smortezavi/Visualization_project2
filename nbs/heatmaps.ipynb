```python
data = pd.read_csv('../data/final-data-commaremoved.csv')
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data.head()
```

```python
prices = data[['Date','dow_price','eur_price','bit_price','gold_price']]
```

```python
prices = prices.set_index("Date")
```

```python
prices = prices.iloc[::-1]
```

```python
pct_change = prices.pct_change()
```

```python
pct_change
```

```python
markets = ['DJIA','EURO','BIT','GOLD']
#py.iplot(data, filename='pandas-heatmap')
data = [
    go.Heatmap(
        z=pct_change.values.tolist(),
        x=markets,
        y=pct_change.index.values,
        colorscale='spectral',
    )
]

layout = go.Layout(
    title='Percent Change in the Markets',
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='datetime-heatmap')
```
